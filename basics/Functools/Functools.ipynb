##### Crie uma função chamada fibonacci que utilize lru_cache do módulo functools para calcular o n-ésimo termo da sequência de Fibonacci. Essa função deve ser recursiva e, graças ao cache, evitar recálculos desnecessários.

```python
@lru_cache
def fibonacci(n: int):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(n=4)
```

```python
fibonacci(20)
```

```python
fibonacci(50)
```

##### Crie uma função chamada verificar_tipo que utilize singledispatch do módulo functools para imprimir uma mensagem diferente dependendo do tipo do argumento fornecido.

##### Crie uma função decoradora chamada decorador_mensagem que utilize wraps do módulo functools para adicionar uma mensagem ao redor da saída de uma função.

```python
from functools import wraps

def decora_msg(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print('Iniciando')
        resultado = func(*args, **kwargs)
        print('Finalizando')
        return resultado
    return wrapper
```

```python
# aqui eu decorei minha função, "embrulhei" com a função wrapper então eu entro na função faço a função wrapper e depois faço a minha func
@decora_msg
def foobar(x, y):
    return x + y

print(foobar(2,2))
```

##### Crie uma classe chamada Pessoa com os atributos nome e idade. Utilize total_ordering do módulo functools para criar os métodos de comparação (__lt__, __le__, __eq__, __ne__, __gt__ e __ge__) com base na idade das pessoas.



```python
from dataclasses import dataclass

@dataclass
class Pessoa:
    name: str = 'default'
    idade: int = 999
    
    # __gt__: corresponde ao >
    def __gt__(self, other):
        return self.idade > other.idade
```

```python
luiz = Pessoa(name='Luiz', idade=26)
maria = Pessoa(name='Maria', idade=23)
```

```python
luiz.idade
```

```python
luiz > maria
```

```python
maria > luiz
```
